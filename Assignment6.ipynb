```python
# Deep Thought 
while True: 
    user_input = input("What is the answer to the Great Question of Life, the Universe, and Everything? (Type 'q' to quit): ") 
    
    if user_input.lower() in ['42', 'forty-two', 'forty two']: 
        print("Yes")  
    elif user_input.lower() == 'q': 
        print("Thanks for trying. Bye!") 
        break 
    else: 
        print("No") 
```

```python
# Playback speed 
while True: 
    user_input = input("Say something please (Q|q to quit): ") 
     
    if user_input.lower() == 'q': 
        print("Bye") 
        break 
    
    slowed_down_text = user_input.replace(' ', '...') 
    print(slowed_down_text) 
```

```python
# Database and Grades 1 
database = {} 

while True: 
    name = input("Enter your name (q|Q to quit): ")
    
    if name.lower() == 'q': 
        print("Bye!") 
        break 
    
    if name in database: 
        password = input("Enter your password: ")
        while password != database[name]['password']:  
            password = input("Incorrect password. Please enter the correct password: ") 
        
        print("Welcome back, {}! Your grades are:".format(name)) 
        for subject, grade in database[name]['grades'].items(): 
            print("{}: {}".format(subject, grade)) 
    else: 
        password = input("Create a password: ") 
        grades = {} 
        for subject in ['Math', 'Physics', 'Geography', 'Chemistry']: 
            grade = int(input("Enter your grade for {}: ".format(subject))) 
            grades[subject] = grade 
        database[name] = {'password': password, 'grades': grades} 
        print("Account created successfully for {}. Welcome!".format(name)) 
```

```python
# Database and Grades 2 
student_scores_1 = { 
    "Einstein": ["5/5", "50/50", "10/10", "10/10"], 
    "Jordan": ["4/8", "50/57", "7/10", "10/18"], 
    "Tesla": ["8/10", "22/25", "3/5", "5/5"], 
    "Curie": ["3/3", "20/20"] 
} 

student_scores_2 = { 
    "Hawking": ["10/10"],
    "Newton": ["30/30"], 
    "Gandhi": ["100/100"], 
    "Cruise": ["1/1"] 
} 

student_scores_3 = { 
    "Turing": ["10/10", "2/4"], 
    "Ali": ["7/9", "2/3"] 
} 

who_passed_1 = [] 
who_passed_2 = [] 
who_passed_3 = [] 

for name, scores in student_scores_1.items(): 
    perfect_score = all(score.split('/')[0] == score.split('/')[1] for score in scores)
    if perfect_score: 
        who_passed_1.append(name) 

for name, scores in student_scores_2.items():
    perfect_score = all(score.split('/')[0] == score.split('/')[1] for score in scores) 
    if perfect_score: 
        who_passed_2.append(name) 

for name, scores in student_scores_3.items():
    perfect_score = all(score.split('/')[0] == score.split('/')[1] for score in scores) 
    if perfect_score: 
        who_passed_3.append(name) 

who_passed_1.sort() 
who_passed_2.sort() 
who_passed_3.sort() 

print("who_passed_1 ->", who_passed_1) 
print("who_passed_2 ->", who_passed_2) 
print("who_passed_3 ->", who_passed_3) 
```
